```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_algo_v1 = pd.read_csv("apa4_algo_v1.csv", comment="#")
```

```python
df_algo_v1
```

```python
df_algo_v2 = pd.read_csv("apa4_algo_v2_pop50_latinhypercube.csv", comment="#")
```

```python
df_algo_v2
```

```python
human_corrected_df = pd.read_csv("human_corrected_apa4_dwa5_XUVG.csv", comment="#")
```

```python
human_corrected_df
```

```python
# Combine the dataframes of the two algorithms and the human corrected data into one dataframe

# First, rename the columns of the human corrected dataframe
human_corrected_df = human_corrected_df.rename(columns={"tension": "human_corrected_tension"})
# We have to also rename the tension column in the algorithm dataframes
df_algo_v1 = df_algo_v1.rename(columns={"tension": "algo_v1_tension", "confidence": "algo_v1_confidence"})
df_algo_v2 = df_algo_v2.rename(columns={"tension": "algo_v2_tension", "confidence": "algo_v2_confidence"})
# Next, merge the two algorithm dataframes with the human corrected dataframe
# on the scan_id and dwa_channel columns
df = pd.merge(df_algo_v1, df_algo_v2, on=["scan_id", "dwa_channel", "side", "layer", "apa_channel", "segment", "headboard_num", "datetime"])
df = pd.merge(df, human_corrected_df, on=["scan_id", "dwa_channel", "side", "layer", "apa_channel", "segment", "headboard_num", "datetime"])
```

```python
mask = (df["layer"] == "X") & np.isfinite(df["algo_v1_tension"])
sns.scatterplot(x="human_corrected_tension", y="algo_v1_tension", data=df[mask], label="Algorithm v1")
sns.scatterplot(x="human_corrected_tension", y="algo_v2_tension", data=df[mask], label="Algorithm v2")
# set labels
plt.xlabel("Human Corrected Tension (N)")
plt.ylabel("Algorithm Tension (N)")
plt.legend()
```

```python
mask = (df["layer"] == "U") & np.isfinite(df["algo_v1_tension"])
sns.scatterplot(x="human_corrected_tension", y="algo_v1_tension", data=df[mask], label="Algorithm v1")
sns.scatterplot(x="human_corrected_tension", y="algo_v2_tension", data=df[mask], label="Algorithm v2")
# set labels
plt.xlabel("Human Corrected Tension (N)")
plt.ylabel("Algorithm Tension (N)")
plt.legend()
```

```python
error_v1 = np.abs(df["human_corrected_tension"] - df["algo_v1_tension"])
error_v2 = np.abs(df["human_corrected_tension"] - df["algo_v2_tension"])

v1_requires_intervention = (error_v1 > 0.5) | (~np.isfinite(df["algo_v1_tension"]))
v2_requires_intervention = (error_v2 > 0.5) | (~np.isfinite(df["algo_v2_tension"]))

print("Algorithm v1 requires intervention for {} of {} samples".format(np.sum(v1_requires_intervention), len(df)))
print("Algorithm v2 requires intervention for {} of {} samples".format(np.sum(v2_requires_intervention), len(df)))
```

```python
layer = "X"
layer_mask = df["layer"] == layer
error_v1 = np.abs(df[layer_mask]["human_corrected_tension"] - df[layer_mask]["algo_v1_tension"])
error_v2 = np.abs(df[layer_mask]["human_corrected_tension"] - df[layer_mask]["algo_v2_tension"])
v1_requires_intervention = (error_v1 > 0.5) | (~np.isfinite(df[layer_mask]["algo_v1_tension"]))
v2_requires_intervention = (error_v2 > 0.5) | (~np.isfinite(df[layer_mask]["algo_v2_tension"]))
print("Algorithm v1 requires intervention for {} of {} samples".format(np.sum(v1_requires_intervention & (df["layer"] == layer)), np.sum(df["layer"] == layer)))
print("Algorithm v2 requires intervention for {} of {} samples".format(np.sum(v2_requires_intervention & (df["layer"] == layer)), np.sum(df["layer"] == layer)))
```

```python
layer = "X"
layer_mask = df["layer"] == layer
error_v1 = np.abs(df[layer_mask]["human_corrected_tension"] - df[layer_mask]["algo_v1_tension"])
error_v2 = np.abs(df[layer_mask]["human_corrected_tension"] - df[layer_mask]["algo_v2_tension"])
# plot cumulative samples below error as a function of the error
sns.lineplot(x=np.sort(error_v1), y=np.arange(len(error_v1)), label="Algorithm v1")
sns.lineplot(x=np.sort(error_v2), y=np.arange(len(error_v2)), label="Algorithm v2")
# set labels
plt.xlabel("Absolute Error (N)")
plt.ylabel("Cumulative Samples")
plt.xlim((-0.01, 1))
```
